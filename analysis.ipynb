```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("arduino_float_data.csv")
```

```python
df
```

```python
df.rename(columns={'X (g)':'Acc_x', 'Y (g)':'Acc_y', 'Z (g)':'Acc_z'}, inplace=True)
```

```python
plt.plot(df[["Acc_x"]], label='Acc_x')
plt.plot(df[["Acc_y"]], label='Acc_y')
plt.plot(df[["Acc_z"]], label='Acc_z')
plt.legend(loc='best')
```

```python
df['Module'] = np.sqrt(df['Acc_x']**2 + df['Acc_y']**2 + df['Acc_z']**2)
plt.plot(df[["Module"]], label='Module')
plt.legend()
```
