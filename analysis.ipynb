```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("capture2.csv")
```

```python
df
```

```python
df.rename(columns={'5.92':'Acc_x', '-0.27':'Acc_y', '7.41':'Acc_z'}, inplace=True)
```

```python
plt.plot(df[["Acc_x"]], label='Acc_x')
plt.plot(df[["Acc_y"]], label='Acc_y')
plt.plot(df[["Acc_z"]], label='Acc_z')
plt.legend(loc='best')
```

```python
df['Module'] = np.sqrt(df['Acc_x']**2 + df['Acc_y']**2 + df['Acc_z']**2)
plt.plot(df[["Module"]], label='Module')
plt.legend()
```
